```python
from numpy import array
import matplotlib.pyplot as plt
def plot_hist(feature, bins = 20):
    x1 = array(dead[feature].dropna())
    x2 = array(survived[feature].dropna())
    plt.hist([x1, x2], label=['Victime', 'Survivant'], bins = bins)#, color = ['', 'b']
    plt.legend(loc = 'upper left')
    plt.ylabel ("Nb. de Pesonnes")
    plt.title('distribution relative de %s' %feature)
    plt.show()
```

```python
child_age = 18
def get_person(passenger):
    age, sex = passenger
    if (age < child_age):
        return 'child'
    elif (sex == 'female'):
        return 'female_adult'
    else:
        return 'male_adult'
```

```python
train = pd.concat([train, pd.DataFrame(train[['Age', 'Sex']].apply(get_person, axis=1), columns=['Person'])],axis=1)
```

```python
def get_type_cabine(cabine):
    # Use a regular expression to search for a title. 
    cabine_search = re.search('\d+', cabine)
    # If the title exists, extract and return it.
    if cabine_search:
        num = cabine_search.group(0)
        if np.float64(num) % 2 == 0:
            return '2'
        else:
            return '1'
    return '0'
train["CabinType"] = train["Cabin"].fillna(" ")

train["CabinType"] = train["CabinType"].apply(get_type_cabine)
```

```python
from sklearn.model_selection import cross_val_score
def compute_score(clf, X, y):
    xval = cross_val_score(clf, X, y, cv = 5)
    print("Accuracy: %0.3f (+/- %0.2f) [%s]" % (xval.mean()*100, xval.std()*100, 'Cross Validation'))
    
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear')
compute_score(lr, X, target)
```

```python
features = ['Fare', 'SibSp', 'Parch', 'Pclass']
def parse_model_1(X):
    X = X[features]
    class_dummies = pd.get_dummies(X['Pclass'], prefix='split_Pclass' )
    X = X.join(class_dummies)
    to_del = ['Pclass']
    for col in to_del : del X[col]
    return X
```

```python
X = parse_model_1(train.copy())
```

```python
X.info()
```

```python
lr = LogisticRegression(solver='liblinear')
compute_score(lr, X, target)
```

```python
lr = LogisticRegression(solver='liblinear')
def coef(clf, X, y):
    clf.fit(X, target)
    print(clf.coef_)
```

```python
features = ['Fare', 'SibSp', 'Parch', 'Pclass', 'Sex']
def parse_model_2(X):
    X = X[features]
    to_dummy = ['Pclass', 'Sex']
    for dum in to_dummy :
        split_temp = pd.get_dummies(X[dum], prefix='split_'+ dum)
        for col in split_temp :
            X[col] = split_temp[col]
        del X[dum]
    return X
```
