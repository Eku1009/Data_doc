### Q1)Autre méthode pour gérer les valeurs manquantes de âge

```python
#Pour gerer autrement  les valeurs manquantes de age  nous commençons par introduire une nnouvelle variable AgeV avec
#des valeurs manquantes. On part importe à nouveaux.

base_train = pd.read_csv('train.csv', sep = ',')
base_test =  pd.read_csv('test.csv', sep = ',')
full2 = pd.concat([base_train,base_test])
full2.set_index('PassengerId',inplace=True,drop=True)
full['AgeV']=full2['Age']
full['Cabin2']=full2['Cabin']

#On utilise ici un algorithme de prediction  pour prédire les valeurs manquantes de âge en prenant comme target  
#ces valeurs existantes.

#Fonction pour avoir colonne catégorisée de sex

def get_cat_sex(sexe):
    if sexe=='female':
        return 1
    else:
        return 0
full['Cat_Sex']=full['Sex'].apply(get_cat_sex)

classers = ['AgeV','Fare','Parch','Pclass','SibSp','TitleCat','CabinCat', 'Cat_Sex','EmbarkedCat', 'FamilySize', 
            'NameLength']
X_full_train = full[classers][full['AgeV'].notnull()] 
Y_full_target = X_full_train['AgeV'].values #full['AgeV'][full['AgeV'].notnull()].values
del X_full_train['AgeV']
X_full_test = full[classers][full['AgeV'].isnull()]
del X_full_test['AgeV']
# Ici nous choissisons l'arbre de décision

from sklearn.ensemble import ExtraTreesRegressor
etr = ExtraTreesRegressor(n_estimators=200)
etr.fit(X_full_train,np.ravel(Y_full_target))
age_preds = etr.predict(X_full_test)
full['AgeV'][full['AgeV'].isnull()]= age_preds


#Ce résume de full montre que la variable AgeV est bien imputer. exemple de l'individu 6
full.head(10)
```

### Q4) Introduisons l'arbre de décision 

```python
#4) Injectons la nouvelle variable AgeV dont les valeurs manquantes ont été imputées dans le features 
train = full[:891]
features = ['Fare', 'SibSp', 'Parch', 'Pclass', 'Sex','TitleCat', 'FamilySize', 'AgeV', 'CabinType', 'Children', 'EmbarkedCat']
def parse_model_6(X):
    X = X[features]
    to_dummy = ['Pclass', 'Sex', 'TitleCat', 'FamilySize', 'CabinType', 'Children', 'EmbarkedCat']
    for dum in to_dummy :
        split_temp = pd.get_dummies(X[dum], prefix='split_'+ dum)
        for col in split_temp :
            X[col] = split_temp[col]
        del X[dum]
    return X

Y = parse_model_6(full.copy())
X=Y[0:891]
```
